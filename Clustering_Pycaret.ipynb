```python
from pycaret.utils import version
version()
```

```python
from pycaret.datasets import get_data
dataSets = get_data('index')
```

```python
wholesale_data = get_data("wholesale")
```

```python
wholesale_data.to_csv("wholesale_data.csv")
from google.colab import files
```

**kmeans clustering**

1.  without preprocessing



```python
from pycaret.clustering import *
setup(data = wholesale_data , verbose = False)
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  x= create_model('kmeans',num_clusters = i)
```

**using z-score normalization**

```python
setup(data = wholesale_data ,  verbose = False , normalize=True , normalize_method='zscore')
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6,7]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters = i)
```

**using transformation**

```python
setup(data = wholesale_data ,  verbose = False , transformation=True , transformation_method='yeo-johnson')
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters = i)
```

**using pca**

```python
setup(data =wholesale_data , verbose=False , pca= True , pca_method='linear')
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters = i)
```

**Transformation + Normalization**

```python
setup(data = wholesale_data , normalize = True , normalize_method='zscore' ,
                             transformation =True , transformation_method= 'yeo-johnson', verbose = False)
plot_model(create_model('kmeans'), plot = 'elbow')

c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters=i)
```

**Transformation + Normalization + PCA**

```python
setup(data = wholesale_data , transformation= True , transformation_method='yeo-johnson',
                             normalize = True , normalize_method= 'zscore',
                             pca=True , pca_method='linear' , verbose=False)
plot_model(create_model('kmeans'), plot = 'elbow')


c= [3,4,5,6]
for i in c:
  print("For number of clusters = ",i)
  create_model('kmeans',num_clusters=i)
```
